# CONSEGNA
Realizzare un chatbot generalista. Ossia un programma Python che chieda all'utente "cosa vuoi?" e generi una risposta tramite un LLM gratuito.
Come LLM gratuito è stato visto Ollama.

```python
import ollama
```

```python
def init_storico():
    messages = [
        {
            "role": "system",
            "content": "Sei un chatbot italiano."
        }
    ]

    return messages
```

```python
def input_utente(storico):
    """
    Chiede all'utente un input, lo pulisce e gestisce il comando di uscita.
    """
    
    user_input = input("User: ").strip()
    input_lower = user_input.lower()
    if input_lower == "stop":
        return None
    storico.append({"role": "user", "content": user_input})
    return storico
    
```

```python
def chiedi_ad_ollama(storico):
    risposta = ollama.chat(
        model = "qwen3.5:4b",
        messages = storico
    )
    testo_risposta = risposta["message"]["content"]
    return testo_risposta
```

```python
def main():
    print("Chatbot locale avviato. Inserisci 'stop' per terminare.\nCome posso aiutarti?")

    storico = init_storico()
    
    while True:
        storico = input_utente(storico)
        if storico is None:
            break
        risposta = chiedi_ad_ollama(storico)
        print(f"Chat: {risposta}\n")
        storico.append({"role": "assistant", "content": risposta})
```

```python
if __name__ == "__main__":
    app = main()
```

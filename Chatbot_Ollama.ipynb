```python
import ollama
```

Si importa ollama per comunicare in maniera agevolata con l' API esposta da Ollama.

```python
class Chatbot:
    """
    Classe che rappresenta un semplice chatbot basato su modello locale Ollama.

    Attributi:
        storico (list): lista di messaggi che rappresenta la conversazione.
                        Ogni messaggio è un dizionario con chiavi 'role' e 'content'.
        modello (str): nome del modello utilizzato per generare le risposte.
    """
    
    def __init__(self):
        self.storico = [
            {
                "role": "system",
                "content": "Sei un chatbot italiano."
            }
        ]
        self.modello = "qwen3.5:4b"

    def chiedi(self):
        """
        Gestisce una singola interazione con la chat: legge input, aggiorna lo storico,
        interroga il modello e restituisce la risposta.
        
        Returns:
            str | None
        """
        self.user_input = input_utente()
        if self.user_input is None:
            return None
        self.storico.append({"role": "user", "content": self.user_input})
        self.risposta = ollama.chat(
            model = self.modello,
            messages = self.storico
        )
        self.testo_risposta = self.risposta["message"]["content"]
        self.storico.append({"role": "assistant", "content": self.testo_risposta})
        return self.testo_risposta
```

```python
def input_utente():
    """
    Chiede all'utente un input, lo pulisce e gestisce il comando di uscita.
    """
    user_input = input("User: ").strip()
    input_lower = user_input.lower()
    if input_lower == "stop":
        return None
    return user_input
    
```

```python
def main():
    """ Definisce il flusso generale del programma. """
    ollama = Chatbot()
    print("Chatbot locale avviato. Inserisci 'stop' per terminare.\nCome posso aiutarti?")
    while True:
        risposta = ollama.chiedi()
        if risposta is None:
            break
        print(f"Chat: {risposta}\n")
```

```python
if __name__ == "__main__":
    app = main()
```
